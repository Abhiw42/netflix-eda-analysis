#### Data Cleaning:
* Check for missing values and handle them
appropriately.
* Remove duplicate entries if any.
* Correct any inconsistencies or errors in the data.

```python
# null values in the dataset in % in each column
df.isnull().sum()/len(df)*100
```

```python
# Convert date_added to datetime
df['date_added']= pd.to_datetime(df['date_added'], errors='coerce')

# Extract other features
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month

# Fill missing categorical values with 'Unknown'
df['director']=df['director'].fillna("Unknown")
df['cast']=df['cast'].fillna("Unknown")
df['country']=df['country'].fillna("Unknown")

# Drop rows with missing 'rating' or 'duration'
df.dropna(subset=['rating', 'duration'], inplace=True)
df.drop(columns="date_added",inplace=True)

# Convert 'date_added' to datetime and Extract other features it creates some nan values
df.isnull().sum()/len(df)*100
```

```python
# ffill (Forward Fill) method, It fills missing values with previous available values
df[["year_added","month_added"]]=df[["year_added","month_added"]].fillna(df[["year_added","month_added"]].ffill)
```

```python
df.info()
```

```python
df.isnull().sum()/len(df)*100
```

```python
movie_data = df[df['type'] == 'Movie']
tv_show_data = df[df['type'] == 'TV Show']

tv_show_data["duration"].unique()
```

```python
#Movies have durations like '90 min', extract the numeric part
movie_data['duration_minutes'] = movie_data['duration'].str.extract('(\d+)').astype(float)
## TV Shows have durations like '1 Season' or '2 Seasons', extract the numeric part
tv_show_data['duration_seasons'] = tv_show_data['duration'].str.extract('(\d+)').astype(float)
```

```python
df.info()
```

```python
# Convert 'year_added' to numeric, handling errors
df['year_added'] = pd.to_numeric(df['year_added'], errors='coerce')

# Drop rows with invalid 'year_added' values (if any)
df.dropna(subset=['year_added'], inplace=True)
```

#### Outliers detection and correction

```python
sns.boxplot(movie_data["duration_minutes"])
plt.show()
```

```python
sns.boxplot(tv_show_data["duration_seasons"])
plt.show()
```

```python
movie_data_clean = movie_data.dropna(subset=['duration_minutes'])

# IQR method for movies
Q1_m = movie_data_clean['duration_minutes'].quantile(0.25)
Q3_m = movie_data_clean['duration_minutes'].quantile(0.75)

lower_m = Q1_m - 1.5 * IQR_m
upper_m = Q3_m + 1.5 * IQR_m

# Filter outliers
movie_data_filtered = movie_data_clean[
    (movie_data_clean['duration_minutes'] >= lower_m) &
    (movie_data_clean['duration_minutes'] <= upper_m)
]
```

```python
# Visualization after handling outliers
plt.figure(figsize=(12, 6))
sns.boxplot(movie_data['duration_minutes'])
plt.title('Boxplot of Movie Durations (After Handling Outliers)')
plt.xlabel('Duration (Minutes)')
plt.show()


plt.figure(figsize=(12, 6))
sns.boxplot(tv_show_data['duration_seasons'])
plt.title('Boxplot of Number of Seasons in TV Shows (After Handling Outliers)')
plt.xlabel('Number of Seasons')
plt.show()
```

#### STEP-15: CONTENT EVOLUTION OVER TIME
EXPLORE HOW THE CHARACTERISTICS OF CONTENT (e.g., DURATION, RATINGS)
HAVE EVOLVED OVER THE YEARS.

```python
# Filter movies with numeric duration (in minutes)
duration_over_time = df[df['duration'].str.contains('min', na=False)].copy()
duration_over_time['duration_mins'] =duration_over_time['duration'].str.extract('(\d+)').astype(float)
# Group by release year and calculate average duration
avg_duration_by_year = duration_over_time.groupby('release_year')['duration_mins'].mean()
plt.figure(figsize=(12, 6))
avg_duration_by_year.plot(marker='o', color='teal')
plt.title('Average Movie Duration Over the Years',fontsize=20)
plt.xlabel('Release Year',fontsize=16)
plt.ylabel('Average Duration (minutes)',fontsize=16)
plt.grid(True)
plt.tight_layout()
plt.show()
```

The peak in movie duration between 1960 and 1966 indicates a trend towards longer films, reflecting a period of artistic and narrative
experimentation in cinema.

```python
# Count of ratings per year
rating_year = df.groupby(['release_year', 'rating']).size().unstack(fill_value=0)
plt.figure(figsize=(16, 7))
rating_year[rating_year.columns[:7]].plot(kind='area', stacked=True, cmap='tab20', alpha=0.8)
plt.title('Distribution of Ratings Over the Years',fontsize=20)
plt.xlabel('Release Year',fontsize=16)
plt.ylabel('Number of Titles',fontsize=16)
plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

The increase in G, NC-17, and NR rated movies until 2018, followed by a gradual decline, reflects a peak in diverse content production, with
a shift in focus or regulation after 2018.
